```python
from unsloth import FastLanguageModel, is_bfloat16_supported
import torch

max_seq_length = 2048 # Can increase for longer reasoning traces
lora_rank = 64 # Larger rank = smarter, but slower


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "Qwen/Qwen2.5-3B-Instruct",
    max_seq_length = max_seq_length,
    load_in_4bit = True, # False for LoRA 16bit
    fast_inference = True, # Enable vLLM fast inference
    max_lora_rank = lora_rank,
    gpu_memory_utilization = 0.5, # Reduce if out of memory
)

model = FastLanguageModel.get_peft_model(
    model,
    r = lora_rank, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ], # Remove QKVO if out of memory
    lora_alpha = lora_rank,
    use_gradient_checkpointing = "unsloth", # Enable long context finetuning
    random_state = 3407,
)
```

```python
import re
import os
from datasets import load_dataset, Dataset

SYSTEM_PROMPT = """
You are a compassionate therapist chatbot. Your goal is to provide empathetic, understanding, and helpful responses to users struggling with mental health concerns.
"""

def get_mental_health_questions(split="train") -> Dataset:
    # ✅ Correct way to load a local JSONL file
    #data = load_dataset("json", data_files={"/train_counselchat_clean_grpo.jsonl"}, split=split)
    url="https://raw.githubusercontent.com/sudo-xda/Kartikeswar-2024/main/train_counselchat_clean_grpo.jsonl"
    data = load_dataset("json", data_files={"train": url}, split=split)

    # ✅ Ensure all fields are strings and correctly structured
    data = data.map(lambda x: {
        'prompt': [
            {'role': 'system', 'content': SYSTEM_PROMPT},
            {'role': 'user', 'content': str(x.get('prompt', ''))}  # Ensure prompt is a string
        ],
        'answer': str(x.get('completion', '')),  # Ensure completion is a string
        'topic': str(x.get('topic', ''))  # Ensure topic is a string
    }, remove_columns=data.column_names)  # ✅ Removes old columns to avoid conflicts

    return data

# ✅ Load dataset
dataset = get_mental_health_questions()

# --- Reward Functions for Therapist Chatbot ---

def empathy_reward(completions, **kwargs) -> list[float]:
    """Rewards responses that acknowledge the user's emotions (reduced points)."""
    responses = [completion[0]["content"].lower() for completion in completions]
    empathetic_keywords = ["i understand", "i'm sorry", "that sounds tough", "tell me more", "you're not alone"]

    # Reduced scores: 0.7 for empathetic, 0.3 for neutral
    scores = [0.7 if any(phrase in response for phrase in empathetic_keywords) else 0.3
              for response in responses]
    return scores

def context_awareness_reward(prompts, completions, **kwargs) -> list[float]:
    """Ensures responses are relevant to the user's concern (reduced points)."""
    user_input = prompts[0][-1]['content'].lower()
    response = completions[0][0]["content"].lower()

    # Reduced scores: 0.7 for relevant, 0.3 for irrelevant
    if "anxious" in user_input and "anxious" not in response:
        return [0.3]
    if "sad" in user_input and "sad" not in response:
        return [0.3]
    return [0.7]

def avoid_generic_response(completions, **kwargs) -> list[float]:
    """Penalizes generic responses with reduced penalties."""
    generic_responses = ["i see.", "okay.", "interesting.", "tell me more.", "i understand."]
    responses = [completion[0]["content"].lower() for completion in completions]

    # Inverted scores with reduced values: 0.7 for good, 0.3 for generic
    scores = [0.3 if response in generic_responses else 0.7 for response in responses]
    return scores

def soft_format_reward_func(completions, **kwargs) -> list[float]:
    """Encourages structured responses with lower rewards."""
    pattern = r"(?:i understand|tell me more|you're not alone|i'm here for you)"
    responses = [completion[0]["content"].lower() for completion in completions]

    # Reduced match reward
    matches = [bool(re.search(pattern, r)) for r in responses]
    return [0.3 if match else 0.0 for match in matches]

def topic_relevance_reward(prompts, completions, **kwargs) -> list[float]:
    """Rewards topic relevance with reduced points."""
    user_input = prompts[0][-1]['content'].lower()
    response = completions[0][0]["content"].lower()

    topic_match = re.search(r"topic:\s*(\w+)", user_input)
    topic = topic_match.group(1) if topic_match else None

    if not topic:
        return [0.5]  # Neutral score remains

    # Reduced relevance scoring
    return [0.7 if topic in response else 0.3]

def strict_format_reward_func(completions, **kwargs) -> list[float]:
    """Rewards structured format with lower scores."""
    structured_patterns = [
        r"^(i understand|it sounds like|that must be|i'm here to help|you're not alone)",
        r"(\?|how|what|why|tell me more|can you elaborate)",
    ]

    responses = [completion[0]["content"].lower() for completion in completions]

    scores = []
    for response in responses:
        matches = [bool(re.search(pattern, response)) for pattern in structured_patterns]
        # Reduced tiered scoring
        scores.append(0.7 if all(matches) else 0.4 if any(matches) else 0.1)

    return scores

```

```python
from trl import GRPOConfig, GRPOTrainer
training_args = GRPOConfig(
    use_vllm = True, # use vLLM for fast inference!
    learning_rate = 1e-3,
    adam_beta1 = 0.9,
    adam_beta2 = 0.99,
    weight_decay = 0.1,
    warmup_ratio = 0.1,
    lr_scheduler_type = "cosine",
    optim = "adamw_8bit",
    logging_steps = 1,
    bf16 = is_bfloat16_supported(),
    fp16 = not is_bfloat16_supported(),
    per_device_train_batch_size = 2,
    gradient_accumulation_steps = 2, # Increase to 4 for smoother training
    num_generations = 2, # Decrease if out of memory
    max_prompt_length = 250,
    max_completion_length = 250,
    num_train_epochs = 2, # Set to 1 for a full training run
    max_steps = 250,
    save_steps = 250,
    max_grad_norm = 0.1,
    report_to = "none", # Can use Weights & Biases
    output_dir = "/mnt/c/Users/SOEE/Desktop/Jeevan/Llma3B",
)
```

```python
trainer = GRPOTrainer(
    model=model,
    reward_funcs=[
        empathy_reward,
        context_awareness_reward,
        avoid_generic_response,
        soft_format_reward_func,
        strict_format_reward_func,
        topic_relevance_reward,
        #rouge_reward_func,
        #length_similarity_reward_func,
        #combined_reward,

    ],
    args=training_args,
    train_dataset=dataset,
)

trainer.train()
```
